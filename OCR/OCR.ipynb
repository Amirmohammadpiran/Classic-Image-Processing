### Loading the necessary libraries:

```python
import cv2
from pytesseract import image_to_string
import numpy as np
import matplotlib.pyplot as plt
```

### Loading the image:

```python
image = cv2.imread("car.jpg")
grayimg = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(grayimg, cmap='gray')
plt.show()
```

### Cropping the plate:

```python
cropped = image[125:144,97:177]
plt.imshow(cropped, cmap='gray')
plt.show()
```

### Upscaling:

```python
scale_factor = 2
upscaled_image = cv2.resize(cropped, (cropped.shape[1] * scale_factor, cropped.shape[0] * scale_factor), interpolation=cv2.INTER_CUBIC)
plt.imshow(upscaled_image, cmap='gray')
plt.show()
```

### Contrast improvement:

```python
alpha=3
beta=0

adjusted=cv2.addWeighted(upscaled_image,alpha,np.zeros(upscaled_image.shape, upscaled_image.dtype),0,beta)
plt.imshow(adjusted, cmap='gray')
plt.show()
```

### Testing:

```python
text = image_to_string(adjusted)
print("Extracted text:", text)
```
